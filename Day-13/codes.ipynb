```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([x**2 for x in range (101)])
plt.xlabel('$x$')  # latex
plt.ylabel('$x^2$') 
```

```python
x=np.linspace(0,20,100)
plt.plot(x, np.sin(x), 'green')
plt.plot(x, np.cos(x), 'orange')
plt.plot(x, np.zeros(shape=x.shape), 'red')
```

```python
x = np.linspace(0,20,10)
plt.plot(x,x**0.5,'or')  # o - circle, r - red
plt.plot(x, np.log(x), 'ko') # o - circle k - black
```

```python
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.title('sine graph')

# Without the raw string, backslash (\) would be treated as an escape character
# in a regular Python string.
#\\theta ensures that Python interprets the \ as a literal character and
#Matplotlib interprets the \theta as a LaTeX theta symbol.
plt.xlabel('$\\theta$ in rad ->')

# Alternatively, we can use 'r' before the string to denote a raw string

# plt.xlabel(r'$\theta$ in rad ->')

plt.ylabel(r'$sin(\theta)$ ->')
plt.plot(x, np.sin(x), '--')
plt.plot(x, np.sin(x+5), '-.', c='gray')
```

```python
path = './Output_images/sine_graph.png'
fig.savefig(path, format='png')
```

```python
plt.plot(x, x+0, linestyle='dashed')  # --  
plt.plot(x, x+1, linestyle='solid')  # -
plt.plot(x, x+2, linestyle='dashdot')  # -.
plt.plot(x, x+3, ls='dotted')  # :
plt.title('Linestyles demo')
```

```python
d = np.linspace(0, 10, 40)
plt.plot(d, d, '*')
plt.plot(d, d+1, ',')
plt.plot(d, d+2, '^')
plt.plot(d, d+3, '1')
plt.plot(d, d+4, '2')
plt.plot(d, d+5, '-2')
plt.plot(d, d+6, '--2')
```

```python
plt.plot(x, x, lw=3, label='linear')
plt.plot(x, x**2, label='$x^2$')
plt.plot(x, x**3, label='$x^3$')
plt.plot(x, x**0.5, label='$x^{0.5}$')
plt.legend()
```

```python
x=np.linspace(0,20,100)
plt.plot(x, np.sin(x))
plt.plot(x,np.array([0.75]*x.shape[0]))
```

```python
x = np.linspace(0, 20, 100)
plt.xlim(0.5,2.5)
plt.ylim(0,1.25)
plt.plot(x, np.sin(x))
plt.plot(x, np.array([0.75]*x.shape[0]))
```

```python
#Subplots
plt.figure()

plt.subplot(2, 2, 1)  # row,col,identity
plt.plot(x, np.sin(x))

plt.subplot(2, 2, 2)  # row,col,identity
plt.plot(x, np.cos(x))

plt.subplot(2, 2, 3)  # row,col,identity
plt.plot(x, np.tan(x))

plt.subplot(2, 2, 4)  # row,col,identity
plt.plot(x, np.exp(x))
```
