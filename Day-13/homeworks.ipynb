```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,1,6)
```

```python
powers=[1/20, 1/5, 1/4, 1/3, 1/2, 1 ,2, 3, 4, 5, 20]
powers
```

```python
for power in powers:
    plt.plot(x, x**power)
```

```python
# Edges were sharp
x=np.linspace(0,1,500) #add more points for a smoother curve
```

```python
plt.xlim(0, 1)
plt.ylim(0, 1)

for power in powers:
    plt.plot(x, x**power)
```

```python
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.gca().spines['top'].set_visible(False) # remove top and right spines
plt.gca().spines['right'].set_visible(False)

for power in powers:
    plt.plot(x, x**power)
```

```python
colors = ['#00dcff', '#22bed6', '#2a72ab', '#20467b', '#122d51', '#061222'
          , '#122d51', '#20467b','#2a72ab','#22bed6','#00dcff'] #colors for the lines
```

```python
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.gca().spines['top'].set_visible(False) 
plt.gca().spines['right'].set_visible(False)
plt.gca().set_facecolor('#F5DEB3') 
plt.gcf().set_facecolor('#F5DEB3') # change background color


for (power,color) in zip(powers,colors):
    plt.plot(x, x**power, c=color)
```
